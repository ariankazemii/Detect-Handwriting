### Read and create train and test dataset

```python
import csv
import pandas as pd
import numpy as np

train_dataset = "pendigits.tra" 

X_train, y_train = [], []

with open(train_dataset, 'r') as file:
    csv_reader = csv.reader(file)
    for row in csv_reader:
        values = list(map(int, row))
        label = values.pop()
        X_train.append(values)
        y_train.append(label)
                
X_train = pd.DataFrame(np.array(X_train))
y_train = pd.Series(np.array(y_train), name='label')


test_dataset = "pendigits.tes"

X_test, y_test = [], []

with open(test_dataset, 'r') as file:
    csv_reader = csv.reader(file)
    for row in csv_reader:
        values = list(map(int, row))
        label = values.pop()
        X_test.append(values)
        y_test.append(label)

X_test = pd.DataFrame(np.array(X_test))
y_test = pd.Series(np.array(y_test), name='label')
```

```python
from sklearn.preprocessing import StandardScaler

standard_scaler = StandardScaler()
x_train = standard_scaler.fit_transform(X_train)
x_test = standard_scaler.transform(X_test)
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score, classification_report
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
import seaborn as sns

def evaluate_MLP(X_train, X_test, y_train, y_test, activation, hidden_layer_size):
    mlp = MLPClassifier(hidden_layer_sizes=hidden_layer_size, activation=activation, max_iter=1000, random_state=42)
    mlp.fit(X_train, y_train)

    y_pred = mlp.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    
    print(f"Activation: {activation}, Hidden Layer Size: {hidden_layer_size}")
    print(f"Accuracy: {accuracy * 100:.2f}%")
    print(classification_report(y_test, y_pred))

    cm = confusion_matrix(y_test, y_pred)

    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Greens', xticklabels=mlp.classes_, yticklabels=mlp.classes_)
    plt.title(f"Confusion Matrix for Activation: {activation}, Hidden Layer Size: {hidden_layer_size}")
    plt.xlabel('Predicted')
    plt.ylabel('True')
    plt.show()

activation_functions = ["tanh", "relu", "logistic"]
hidden_layer_sizes = [(64, 64), (128, 128)]

for activation in activation_functions:
    for hidden_size in hidden_layer_sizes:
        evaluate_MLP(X_train, X_test, y_train, y_test, activation, hidden_size)
```
